```python
# Modules imported
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("FinalCleanedDataset.csv")
```

```python
plt.figure(figsize=(15, 10))

# Boxplots of `Price` by key categorical variables
plt.figure(figsize=(15, 10))

for i, var in enumerate(['BodyType', 'FuelType', 'Transmission', 'UsedOrNew'], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=data, x=var, y='Price', palette='Set3')
    plt.title(f'Boxplot of Price by {var}', fontsize=14)
    plt.xticks(rotation=45, fontsize=10)
    plt.xlabel(var, fontsize=12)
    plt.ylabel('Price', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
numerical_vars = ['Price', 'Kilometres', 'EngineCapacity', 'FuelConsumptionPer100km']

# Plot histograms for the numerical variables
plt.figure(figsize=(15, 10))

for i, var in enumerate(numerical_vars, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[var], kde=True, color='skyblue', bins=20)
    plt.title(f'Histogram of {var}', fontsize=14)
    plt.xlabel(var, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Plot bar charts for categorical variables
categorical_vars = ['UsedOrNew', 'Transmission', 'FuelType', 'YearRanges']

plt.figure(figsize=(15, 10))

for i, var in enumerate(categorical_vars, 1):
    plt.subplot(2, 2, i)
    data[var].value_counts().plot(kind='bar', color='coral', alpha=0.8)
    plt.title(f'Bar Chart of {var}', fontsize=14)
    plt.xlabel(var, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Calculate correlation matrix for numerical variables
correlation_matrix = data[numerical_vars].corr()

# Plot heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True, square=True)
plt.title('Heatmap of Correlation Among Numerical Variables', fontsize=14)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Scatter plots for `Price` vs numerical predictors
plt.figure(figsize=(15, 10))

for i, var in enumerate(['Kilometres', 'EngineCapacity', 'FuelConsumptionPer100km'], 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(data=data, x=var, y='Price', alpha=0.6, color='blue')
    plt.title(f'Scatter Plot of Price vs {var}', fontsize=14)
    plt.xlabel(var, fontsize=12)
    plt.ylabel('Price', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Calculate summary statistics for numerical variables
summary_stats = data[numerical_vars].describe().T
summary_stats['CV'] = (summary_stats['std'] / summary_stats['mean']) * 100  # Coefficient of variation

# Add quartiles
summary_stats['Q1'] = data[numerical_vars].quantile(0.25)
summary_stats['Q3'] = data[numerical_vars].quantile(0.75)

summary_stats
```
